```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ^STOXX50E
df = pd.read_csv('euro_stoxx50.csv')
df
```

```python
# Rename date column:
df.rename(columns={"Price": "Date"}, inplace=True)
```

```python
df.drop([0, 1], axis=0, inplace=True)  # Delete rows with index 0 and 1.
```

```python
df
```

```python
print(df.dtypes)
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d")
```

```python
# Convert numeric columns
cols_numeric = ["Close", "High", "Low", "Open", "Volume"]
df[cols_numeric] = df[cols_numeric].apply(pd.to_numeric, errors="coerce")
```

```python
print(df.dtypes)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["Date"], df["Close"], label="Close", color="blue")
plt.plot(df["Date"], df["Open"], label="Open", color="orange", alpha=0.6)
plt.title("EURO STOXX 50 - Temporal Series")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
